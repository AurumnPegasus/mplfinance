# HOLLOW and FILLED CANDLESTICKS


### To best understand Hollow and Filled Candlesticks, we compare them to Tradiational Candlesticks:

---

# Traditional Candlestick Charts:
- Traditional candlestick charts **contain only TWO Candle Colors**.  These two colors represent whether the 
  - Close is GREATER than the Open (**up** color), or the
  - Close is LESS than the Open (**down** color).
- **Example** Color Combinations:

| <h2>`Black (Close > Open)`<br><br>`Red   (Close < Open)`</h2> | <h2>`White (Close > Open)`<br><br>`Black (Close < Open)`</h2> | <h2>`Green (Close > Open)`<br><br>`Red   (Close < Open)`</h2> |
| :---- | :---- | :---- |
|  |  |  |

---
# Hollow and Filled Candlestick Charts:

### Hollow and Filled Candlesticks differ from Traditional Candlesticks in two imporant ways:

**1.** The **TWO** COLORS ***now*** represent the ***Change (up/down) FROM the PREVIOUS CLOSE*** (*not* from the Open).

**2.** The  **Change from the Open** is instead represented by whether or not the candle body is 
- **HOLLOW**: Close **UP**&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; from the **Open**, or
- **FILLED**: &nbsp;&nbsp;&nbsp;Close **DOWN** from the **Open**

---
## EXAMPLES:


## Hollow and Filled, using Black/Red Candle Colors:

| <h3>BLACK:<li>Up from Previous Close</li><br>FILLED:<li>Down from Open</li></h3> | <h3>RED:<li>Down from Previous Close</li><br>FILLED:<li>Down from Open</li></h3> | <h3>BLACK:<li>Up from Previous Close</li><br>HOLLOW:<li>Up from Open</li></h3> | <h3>RED:<li>Down from Previous Close</li><br>HOLLOW:<li>Up from Open</li></h3> |

|  |  |  |  |

## A NOTE of WARNING:
- Some Candlestick Styles have a Edge and Wick Color that *differs* from the Candle Body Color.  
- This may be done to provide contrast between the candle color and the background or face color of the plot.
<br><br>
- When this happens ***it may appear that some of the candles are hollow*** when really we are looking at a ***Traditional*** candlestick plot but with some candles ***outlined*** to contrast with the face color of the plot.
<br><br>
- For example, "Classic" candlestick colors are White and Black, on a White Background, so in order to be able to see the White Candles (against a white background) they are given a black outline, as follows:
